```python
import requests
import json
import pandas as pd

base_url = "https://tienda.mercadona.es/api/categories/"
headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
    }


all_products=[]
```

```python
response = requests.get(base_url, headers=headers)

```

```python
response.raise_for_status()
```

```python
categories_data = response.json().get('results', [])

        
```

```python
categories_data 
```

```python
cookies = {'customer_postal_code': "28001"}
```

```python
for root in categories_data:
            print(f"guardando sección: {root['name']}")
            
            for sub in root.get('categories', []):
                detail_res = requests.get(f"{base_url}{sub['id']}/", headers=headers, cookies=cookies)
                
                if detail_res.status_code == 200:
                    sub_data = detail_res.json()
                    
                    for level3 in sub_data.get('categories', []):
                        for item in level3.get('products', []):
                            price_info = item.get('price_instructions') or {} 
                            all_products.append({
                                "id": item.get('id'),
                                "nombre": item.get('display_name'),
                                "precio_unidad": price_info.get('unit_price'),
                                "precio_bulk": price_info.get('bulk_price'), 
                                "formato": item.get('format'),
                                "categoria": level3.get('name'),
                                "seccion": root.get('name')
                            })

```

```python
with open('mercadona_precios_dataset.json', 'w', encoding='utf-8') as f:
    json.dump(all_products, f, ensure_ascii=False, indent=4)
        
print(f"¡Dataset completado! Se han guardado {len(all_products)} productos en 'mercadona_precios_dataset.json'.")
```

```python
df = pd.read_json('mercadona_precios_dataset.json')

df.to_csv('mercadona_precio_dataset.csv', index=False)
```
